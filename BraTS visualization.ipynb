```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
drive.mount('/content/drive')
filename = '/content/drive/My Drive/Colab Notebooks/data/BRATS2015_Training/HGG/BraTS19_TCIA03_498_1/BraTS19_TCIA03_498_1_t1.nii.gz' 
```

```python
img = nib.load(filename)
```

```python
print(img.shape)
print(img.get_data_dtype() == np.dtype(np.int16))
print(img.affine.shape)
data = img.get_fdata()
print(data.shape)
hdr = img.header
print(hdr)
print(hdr.get_xyzt_units())
```

```python
plt.figure(figsize=(50,50))
for i in range(data.shape[2]):
    plt.subplot(23,8,i+1), plt.imshow(data[:,:,i]), plt.axis('off')
    # use plt.savefig(...) here if you want to save the images as .jpg, e.g.,
plt.show()
```
